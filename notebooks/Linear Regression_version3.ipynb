## Linear Regression Modell

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV, KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.base import clone
from sklearn.model_selection import cross_val_score

# Define a custom function to perform cross-validation with different random states and cv values
def random_search_cv(X, y, random_state_range, cv_range, n_iter=10):
    best_score = -np.inf
    best_params = None
    best_model = None

    for _ in range(n_iter):
        # Randomly select random_state and cv from the given ranges
        random_state = np.random.choice(random_state_range)
        cv_folds = np.random.choice(cv_range)

        # Split the data with the current random_state
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
        
        # Define pipeline with imputation, scaling, and linear regression
        pipeline = Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler()),
            ('regressor', LinearRegression())
        ])

        # Perform cross-validation with the current cv and pipeline
        cv = KFold(n_splits=cv_folds, shuffle=True, random_state=random_state)
        cv_scores = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring='r2')
        mean_cv_score = np.mean(cv_scores)
        
        # Update best parameters if current model performs better
        if mean_cv_score > best_score:
            best_score = mean_cv_score
            best_params = {'random_state': random_state, 'cv': cv_folds}
            best_model = clone(pipeline).fit(X_train, y_train)
            
    return best_model, best_params, best_score

# Define ranges for random_state and cv
random_state_range = range(0, 100, 5)  # e.g., 0, 5, 10, ..., 95
cv_range = range(3, 11)  # e.g., 3, 4, ..., 10

# Run the custom random search
best_model, best_params, best_score = random_search_cv(X, y, random_state_range, cv_range, n_iter=10)

# Display best parameters and score
print("Best parameters found:", best_params)
print("Best cross-validated R² score:", best_score)

# Evaluate on the test set using the best model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=best_params['random_state'])
y_pred = best_model.predict(X_test)

# Calculate error metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Test Set Mean Squared Error (MSE): {mse}")
print(f"Test Set Mean Absolute Error (MAE): {mae}")
print(f"Test Set R-squared (R²): {r2}")

# Extract and print the coefficients from the model
coefficients = best_model.named_steps['regressor'].coef_
intercept = best_model.named_steps['regressor'].intercept_

# Display coefficients with feature names
print("Model Intercept:", intercept)
print("Model Coefficients:")
for feature, coef in zip(X.columns, coefficients):
    print(f"{feature}: {coef}")
```
